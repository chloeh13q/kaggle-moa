```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
train_x = pd.read_csv('data/train_features.csv', index_col=0)
train_y = pd.read_csv('data/train_targets_scored.csv', index_col=0)
train_op = pd.read_csv('data/train_targets_nonscored.csv', index_col=0)
```

```python
g_cols = [col for col in train_x.columns if col.startswith("g-")]
c_cols = [col for col in train_x.columns if col.startswith("c-")]
```

```python
pca = PCA()
pca.fit(train_x.loc[:, g_cols])
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

pca = PCA()

```python
pca = PCA()
pca.fit(train_x[train_x['cp_type'] == 'trt_cp'].loc[:, c_cols])
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
pca_g = PCA()
pca_g.fit(train_x.loc[:, g_cols])

pca_c = PCA()
pca_c.fit(train_x.loc[:, c_cols])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6), dpi=150)
ax1.set_title('Scree Plot (Gene Expression Features)')
ax1.bar(range(1, len(pca_g.explained_variance_ratio_) + 1), pca_g.explained_variance_ratio_)
ax1.plot(np.cumsum(pca_g.explained_variance_ratio_), c='red')
ax1.set(xlabel='# of principal components', ylabel='Percentage of variance explained')

ax2.set_title('Scree Plot (Cell Viability Features)')
ax2.bar(range(1, len(pca_c.explained_variance_ratio_) + 1), pca_c.explained_variance_ratio_)
ax2.plot(np.cumsum(pca_c.explained_variance_ratio_), c='red')
ax2.set(xlabel='# of principal components', ylabel='Percentage of variance explained')

plt.savefig('scree_plot.png')
```
